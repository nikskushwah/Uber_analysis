```python
%pylab inline
import pandas as pd
import seaborn as sn
```

```python
data = pd.read_csv("uber-raw-data-apr14.csv")
```

```python
data.tail()
```

```python
data['Date/Time'] = data['Date/Time'].map(pd.to_datetime)
```

```python
def get_dom(dt):
    return dt.day

data['dom'] = data['Date/Time'].map(get_dom)
```

```python
data.tail()
```

```python
def get_weekday(dt):
    return dt.weekday()

data['weekday'] = data['Date/Time'].map(get_weekday)
```

```python
def get_hour(dt):
    return dt.hour

data['hour'] = data['Date/Time'].map(get_hour)
```

```python
data.tail()
```

```python
hist(data.dom, bins = 30, rwidth = 0.8, range = (0.5,30.5))
xlabel("date of the month")
ylabel("frequency")
```

```python
#for k,rows in data.groupby('dom'):
 #   print((k,len(rows)))
    
    
def count_rows(row):
    return len(row)

by_date = data.groupby('dom').apply(count_rows)
by_date
```

```python
plot(by_date)
```

```python
bar(range(1,31), by_date)
```

```python
by_date_sorted = by_date.sort_values()
by_date_sorted
```

```python
bar(range(1,31), by_date_sorted)xticks(range(1,31), by_date_sorted.index)
;
xticks(range(1,31), by_date_sorted.index)
;
```

```python
hist(data.hour, bins = 24, rwidth = 0.8, range = (0.5,24))
xlabel("Hour")
ylabel("frequency")
```

```python
hist(data.weekday, bins = 7, rwidth = 0.8, range = (-0.5,6.5))
xticks(range(7), 'Mon Tue Wed Thur Fri Sat Sun'.split(' '))
;
```
